```python
import pandas as pd
import plotly.express as px
import numpy as np
from scipy.signal import lfilter
```

```python

def plot_in_RCS(robot_log_fname,sensor_log_fname, transformation_matrix_camera_to_rcs, noise = False):
  new = []
  df = pd.read_csv(sensor_log_fname, skiprows=6,usecols=["X.1", "Y.1", "Z.1"]).dropna()
  sensor_cartesian_array = np.array([df["X.1"].array.to_numpy(), df["Y.1"].array.to_numpy(), df["Z.1"].array.to_numpy()])
  for a in sensor_cartesian_array.transpose():
      a = np.append(a, 1)
      new.append(np.dot(transformation_matrix_camera_to_rcs, a.T))
  ufo = pd.DataFrame(new, columns=['X', 'Y', 'Z', 'Scaling'])
  ufo['CS'] = 'sensor_in_RCS'
  fig = px.line_3d(ufo,x='X', y='Y', z='Z',labels='sensor',color='CS')
  my_robot = pd.read_csv(robot_log_fname)  
  my_robot['CS']='Robot(RCS)'
  fig.add_traces(px.line_3d(my_robot,x='X',y='Y',z='Z',color='CS',color_discrete_sequence=['indianred']).data)
  fig.show()
```
